# **Vision Transformer**
Vision Transformer dalam AI berperan layaknya otak dalam menganalisis gambar. AI menerima gambar wajah manusia, kemudian memotongnya menjadi bagian-bagian kecil (patch). Masing-masing patch dipelajari dengan cermat, mengamati detail seperti warna, tekstur, dan bentuk.

Informasi dari patch-patch ini kemudian digabungkan dan diolah oleh jaringan saraf tiruan, bagaikan detektif yang mencari pola tersembunyi. Pola-pola ini terkait dengan kondisi kulit yang berbeda, seperti jerawat, kemerahan, dan kantung mata.

Jaringan saraf tiruan kemudian membuat kesimpulan. Ia menentukan kategori yang paling mungkin untuk kondisi kulit di gambar, beserta tingkat probabilitasnya. Hasil ini disajikan kepada pengguna, membantu mereka memahami kondisi kulit mereka dengan lebih baik.

Namun, perlu diingat bahwa AI ini adalah alat bantu, bukan dokter. Konsultasi dengan dokter atau profesional kesehatan selalu direkomendasikan untuk diagnosis dan penanganan yang tepat.

1. Menerima gambar wajah manusia sebagai input. Gambar dipastikan berkualitas baik dengan pencahayaan yang cukup dan resolusi yang memadai untuk analisis visual.
2. Vision Transformer memecah gambar menjadi patch-patch kecil dan menganalisis fitur dan pola di setiap patch.
3. Hubungan antar patch kemudian dipelajari untuk memahami konteks gambar secara keseluruhan.

# **3. Preprocessing**

Membagi dataset menjadi set pelatihan, validasi, dan pengujian.

```python
images_path = [None] * len(IMAGE_PATH_LIST)
labels = [None] * len(IMAGE_PATH_LIST)

for i,image_path in enumerate(IMAGE_PATH_LIST):
    images_path[i] = image_path
    labels[i] = image_path.parent.parent.stem
    
df_path_and_label = pd.DataFrame({'path':images_path, 
                                  'label':labels})
df_path_and_label.head()
```

```python
SEED = 123

df_train, df_rest = tts(df_path_and_label, 
                        test_size = 0.3, 
                        random_state = SEED, 
                        stratify = df_path_and_label["label"])

df_val, df_test = tts(df_rest, 
                      test_size = 0.5, 
                      random_state = SEED, 
                      stratify = df_rest["label"])
```

Mendefinisikan pemetaan label kelas ke nilai numerik.

```python
label_map = dict(zip(classes, range(0, len(classes))))
label_map
```

Mendefinisikan transformasi default untuk preprocessing gambar menggunakan model Vision Transformer (ViT).

```python
weights = ViT_B_16_Weights.DEFAULT
auto_transforms = weights.transforms()
auto_transforms
```

Mendefinisikan kelas dataset kustom untuk menangani dataset.

```python
class CustomDataset(Dataset):
    def __init__(self, df:pd.DataFrame, transforms, label_map:dict):
        self.df = df
        self.transforms = transforms
        self.label_map = label_map
        
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
        df_new = self.df.copy()
        df_new = df_new.reset_index(drop = True)
        df_new["label"] = df_new["label"].map(self.label_map)
        image_path = df_new.iloc[idx, 0]
        image = Image.open(image_path).convert("RGB")
        image = self.transforms(image)
        label = df_new.iloc[idx, 1]
        
        return image,label
```

Membuat instansi dataset kustom untuk set pelatihan dan validasi.

```python
train_dataset = CustomDataset(df_train, auto_transforms, label_map)
valid_dataset = CustomDataset(df_val, auto_transforms, label_map)
```

Membuat pengambil data untuk set pelatihan dan validasi untuk memuat data secara efisien dalam batch.

Mendefinisikan fungsi loss dan optimizer.

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr = 0.01)
```

Mendefinisikan fungsi untuk satu langkah pelatihan.

```python
def train_step(model:torch.nn.Module, 
               dataloader:torch.utils.data.DataLoader, 
               loss_fn:torch.nn.Module, 
               optimizer:torch.optim.Optimizer):
    
    model.train()
    
    train_loss = 0.
    train_accuracy = 0.
    
    for batch,(X,y) in enumerate(dataloader):
        X,y = X.to(device), y.to(device)
        optimizer.zero_grad()
        y_pred_logit = model(X)
        loss = loss_fn(y_pred_logit, y)
        train_loss += loss.item()
        
        loss.backward()
        optimizer.step()
        
        y_pred_prob = torch.softmax(y_pred_logit, dim = 1)
        y_pred_class = torch.argmax(y_pred_prob, dim = 1)
        train_accuracy += accuracy_score(y.cpu().numpy(), 
                                         y_pred_class.detach().cpu().numpy())
        
    train_loss = train_loss/len(dataloader)
    train_accuracy = train_accuracy/len(dataloader)
    
    return train_loss, train_accuracy
```

Mendefinisikan fungsi untuk menyimpan checkpoint model.

```python
def save_checkpoint(filename, model, loss, epoch, optimizer, metric):
    state = {"filename":filename, 
             "model":model.state_dict(), 
             "loss":loss, 
             "epoch":epoch, 
             "optimizer":optimizer.state_dict(), 
             "metric":metric}
    
    torch.save(state, filename)
```

Mendefinisikan fungsi untuk satu langkah validasi.

```python
def valid_step(model:torch.nn.Module, 
               dataloader:torch.utils.data.DataLoader, 
               loss_fn:torch.nn.Module):
    
    model.eval()
    
    valid_loss = 0.
    valid_accuracy = 0.
    
    with torch.inference_mode():
        for batch,(X,y) in enumerate(dataloader):
            X,y = X.to(device), y.to(device)
            y_pred_logit = model(X)
            loss = loss_fn(y_pred_logit, y)
            valid_loss += loss.item()
            
            y_pred_prob = torch.softmax(y_pred_logit, dim = 1)
            y_pred_class = torch.argmax(y_pred_prob, dim = 1)
            
            valid_accuracy += accuracy_score(y.cpu().numpy(), y_pred_class.detach().cpu().numpy())
            
    valid_loss = valid_loss/len(dataloader)
    valid_accuracy = valid_accuracy/len(dataloader)
    
    return valid_loss, valid_accuracy
```

Mendefinisikan loop utama pelatihan.

```python
def train(model:torch.nn.Module, 
          train_dataloader:torch.utils.data.DataLoader, 
          valid_dataloader:torch.utils.data.DataLoader, 
          loss_fn:torch.nn.Module, 
          optimizer:torch.optim.Optimizer, 
          epochs:int = 10):
    
    results = {"train_loss":[], 
               "train_accuracy":[], 
               "valid_loss":[], 
               "valid_accuracy":[]}
    
    best_valid_loss = float("inf")
    
    for epoch in tqdm(range(epochs)):
        train_loss, train_accuracy = train_step(model = model, 
                                                dataloader = train_dataloader, 
                                                loss_fn = loss_fn, 
                                                optimizer = optimizer)
        
        valid_loss, valid_accuracy = valid_step(model = model, 
                                                dataloader = valid_dataloader, 
                                                loss_fn = loss_fn)
        
        if valid_loss < best_valid_loss:
            best_valid_loss = valid_loss
            file_name = "best_model.pth"
            save_checkpoint(file_name, model, best_valid_loss, epoch, optimizer, valid_accuracy)
            
        print(f"Epoch: {epoch + 1} | ", 
              f"Train Loss: {train_loss:.4f} | ", 
              f"Train Accuracy: {train_accuracy:.4f} | ", 
              f"Valid Loss: {valid_loss:.4f} | ", 
              f"Valid Accuracy: {valid_accuracy:.4f}")
        
        results["train_loss"].append(train_loss)
        results["train_accuracy"].append(train_accuracy)
        results["valid_loss"].append(valid_loss)
        results["valid_accuracy"].append(valid_accuracy)
        
    return results
```

Melatih model.

```python
# Training!!!
EPOCHS = 10

torch.cuda.manual_seed(SEED)
torch.manual_seed(SEED)

MODEL_RESULTS = train(model.to(device), 
                      train_dataloader, 
                      valid_dataloader, 
                      loss_fn, 
                      optimizer, 
                      EPOCHS)
```

Mendefinisikan fungsi untuk memplot kerugian dan akurasi selama pelatihan.

```python
def loss_metric_curve_plot(model_results:Dict[str,List[float]]):
    
    train_loss = model_results["train_loss"]
    valid_loss = model_results["valid_loss"]
    
    train_accuracy = [float(value) for value in model_results["train_accuracy"]]
    valid_accuracy = [float(value) for value in model_results["valid_accuracy"]]
    
    fig,axes = plt.subplots(nrows = 1, ncols = 2, figsize = (10,4))
    axes = axes.flat
    
    axes[0].plot(train_loss, color = "red", label = "Train")
    axes[0].plot(valid_loss, color = "blue", label = "Valid")
    axes[0].set_title("CrossEntropyLoss", fontsize = 12, fontweight = "bold", color = "black")
    axes[0].set_xlabel("Epochs", fontsize = 10, fontweight = "bold", color = "black")
    axes[0].set_ylabel("Loss", fontsize = 10, fontweight = "bold", color = "black")
    axes[0].legend()
    
    axes[1].plot(train_accuracy, color = "red", label = "Train")
    axes[1].plot(valid_accuracy, color = "blue", label = "Valid")
    axes[1].set_title("Metric of performance: Accuracy", fontsize = 12, fontweight = "bold", color = "black")
    axes[1].set_xlabel("Epochs", fontsize = 10, fontweight = "bold", color = "black")
    axes[1].set_ylabel("Score", fontsize = 10, fontweight = "bold", color = "black")
    axes[1].legend()
    
    fig.tight_layout()
    fig.show()
```

Mendefinisikan fungsi untuk memprediksi label kelas dan merekomendasikan perawatan kulit berdasarkan label yang diprediksi dari gambar yang diunggah.

```python
def recommend_skincare(predicted_label):
    if predicted_label == "acne":
        return "Rekomendasi skincare untuk kulit berjerawat: gunakan produk yang mengandung salicylic acid dan benzoyl peroxide untuk membersihkan dan mengobati jerawat."
    elif predicted_label == "bags":
        return "Rekomendasi skincare untuk mengurangi kantung mata: gunakan krim mata yang mengandung retinol dan vitamin C untuk mengencangkan dan mencerahkan kulit di sekitar mata."
    elif predicted_label == "redness":
        return "Rekomendasi skincare untuk meredakan kemerahan pada wajah: gunakan produk yang mengandung niacinamide dan aloe vera untuk menenangkan dan mengurangi kemerahan pada kulit."
    else:
        return "Tidak ada rekomendasi skincare yang cocok untuk hasil prediksi ini."
    
# Fungsi untuk memprediksi gambar dari PIL.Image
def predict_image_from_pil(image, model, label_map, device):
    # Load and preprocess the image
    transform = ViT_B_16_Weights.DEFAULT.transforms()
    image = transform(image).unsqueeze(0).to(device)

    # Set the model to evaluation mode and perform inference
    model.eval()
    with torch.no_grad():
        y_pred_logit = model(image)
        y_pred_prob = torch.softmax(y_pred_logit, dim=1)
        y_pred_class = torch.argmax(y_pred_prob, dim=1).item()

    # Map the predicted class index to the corresponding label
    inv_label_map = {v: k for k, v in label_map.items()}
    predicted_label = inv_label_map[y_pred_class]

    return predicted_label

# Fungsi untuk memprediksi gambar
def predict_image(change):
    output.clear_output()
    
    # Ambil file yang diupload
    uploaded_file = change['new'][0]
    
    # Baca konten file dan konversi ke gambar
    image_content = uploaded_file['bare face']
    image = Image.open(io.BytesIO(image_content)).convert("RGB")
    
    # Lakukan prediksi pada gambar disini dengan fungsi predict_image
    predicted_label = predict_image_from_pil(image, loaded_model, label_map, device)
    
    # Tampilkan hasil prediksi atau lakukan operasi lainnya
    with output:
        print("Klasifikasi wajah:", predicted_label)
        print("Rekomendasi Perawatan Kulit:", recommend_skincare(predicted_label))

# Melakukan pemantauan pada perubahan yang terjadi di file upload
uploader = widgets.FileUpload(accept='image/*', multiple=False)
uploader.observe(predict_image, names='value')

# Tampilkan widget
display(uploader)
output = widgets.Output()
display(output)
```

Mendefinisikan fungsi untuk memprediksi label kelas dan merekomendasikan perawatan kulit berdasarkan label yang diprediksi dari gambar webcam.

```python
def predict_webcam_image():
    # Inisialisasi detektor wajah
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    cap = cv2.VideoCapture(0)
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Gagal membaca frame dari webcam")
            break

        # Konversi frame ke format yang dapat diproses oleh PIL
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        pil_image = Image.fromarray(frame_rgb)

        # Deteksi wajah
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray_frame, 1.1, 4)

        # Prediksi gambar jika ada wajah yang terdeteksi
        if len(faces) > 0:
            # Misalkan kita menggunakan titik tengah dari wajah pertama yang terdeteksi
            x, y, w, h = faces[0]
            face_pil_image = pil_image.crop((x, y, x + w, y + h))
            predicted_label = predict_image_from_pil(face_pil_image, loaded_model, label_map, device)

            # Rekomendasi skincare berdasarkan hasil prediksi
            skincare_recommendation = recommend_skincare(predicted_label)

            # Tampilkan hasil prediksi dan rekomendasi skincare
            cv2.putText(frame, predicted_label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

            # Split rekomendasi skincare menjadi beberapa baris jika terlalu panjang
            lines = textwrap.wrap(skincare_recommendation, width=40)
            y_offset = 50
            for line in lines:
                cv2.putText(frame, line, (20, y_offset), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                y_offset += 30

        # Tampilkan frame webcam
        cv2.imshow('Webcam', frame)

        # Tunggu tombol 'q' ditekan untuk keluar
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Tutup kamera dan jendela
    cap.release()
    cv2.destroyAllWindows()

# Panggil fungsi untuk memulai prediksi menggunakan webcam
predict_webcam_image()
```
